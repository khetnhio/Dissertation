Some EDA and label distribution checking on the proposed test, validation and train datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/My Drive/Dissertation/pneumonia_full.csv')
```

```python
df.head()
```

```python
#split the data based on the reccomended split
train = df[df['split']=='train']
val = df[df['split']=='validate']
test = df[df['split']=='test']
assert len(df) == len(train)+len(val)+len(test)
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is already loaded and is called df

# Create the histogram
plt.hist(df['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')

# Adding titles and labels
plt.title('Histogram of Pneumonia labels')
# plt.xlabel('Pneumonia Class')
plt.ylabel('Frequency')
plt.xticks([-1, 0, 1])
# Show the plot
plt.show()



# Create the histogram
plt.hist(train['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')

# Adding titles and labels
plt.title('Histogram of Pneumonia labels in the train dataset')
# plt.xlabel('Pneumonia Class')
plt.ylabel('Frequency')
plt.xticks([-1, 0, 1])
# Show the plot
plt.show()



# Create the histogram
plt.hist(val['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')

# Adding titles and labels
plt.title('Histogram of Pneumonia labels in the validation dataset')
# plt.xlabel('Pneumonia Class')
plt.ylabel('Frequency')
plt.xticks([-1, 0, 1])
# Show the plot
plt.show()



# Create the histogram
plt.hist(test['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')

# Adding titles and labels
plt.title('Histogram of Pneumonia labels in the test dataset')
# plt.xlabel('Pneumonia Class')
plt.ylabel('Frequency')
plt.xticks([-1, 0, 1])
# Show the plot
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].hist(df['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')
axes[0, 0].set_title('Histogram of Pneumonia labels on the full dataset')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_xticks([-1, 0, 1])

axes[0, 1].hist(train['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')
axes[0, 1].set_title('Histogram of Pneumonia labels in the train dataset')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_xticks([-1, 0, 1])

axes[1, 0].hist(val['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')
axes[1, 0].set_title('Histogram of Pneumonia labels in the validation dataset')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_xticks([-1, 0, 1])

axes[1, 1].hist(test['pneumonia'], bins=[-1.5, -0.5, 0.5, 1.5], color = 'seagreen',edgecolor='black', align='mid')
axes[1, 1].set_title('Histogram of Pneumonia labels in the test dataset')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_xticks([-1, 0, 1])
plt.tight_layout()
plt.show()
```
